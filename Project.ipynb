```python
import os
folderpath = r"E:\wiki" 
filepaths = [os.path.join(folderpath,name) for name in os.listdir(folderpath)]
articles = []
for path in filepaths:
    with open(path,"r",encoding="ISO-8859-1") as f:
        file = f.readlines()
        articles.append(file)
```

```python
articles
```

```python
articles_str = ' '.join(map(str,articles)) 
```

```python
articles_str
```

```python
import nltk
```

```python
nltk.download()
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.tokenize import word_tokenize
```

```python
text = word_tokenize(articles_str)
data = [word for word in text if not word in stopwords.words()]
```

```python
print(data)
```

```python
data
```

```python
from nltk.stem import WordNetLemmatizer
All=WordNetLemmatizer()
articles_lem=[]
for r in data:
    articles_lem.append(All.lemmatize(r))
print(articles_lem)
```

```python
import re
dataupdate = [ ]
dataupdate= [re.sub('[^a-zA-Z0]+','', _)for _ in articles_lem]
dataupdate
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
#import warnings filter
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
vectorizer=TfidfVectorizer(stop_words='english')
X=vectorizer.fit_transform(dataupdate)
true_k=5
model=KMeans(n_clusters=true_k,init='k-means++', max_iter=100,n_init=1)
model.fit(X)
print("Top terms per cluster:")
order_centroids= model.cluster_centers_.argsort()[:,::-1]
terms= vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i,:10]:
        print(' %s' % terms[ind]),
```

```python
print("Guessing :")
Y = vectorizer.transform(["human york function"])
guessing= model.predict(Y)
print(guessing)
```
